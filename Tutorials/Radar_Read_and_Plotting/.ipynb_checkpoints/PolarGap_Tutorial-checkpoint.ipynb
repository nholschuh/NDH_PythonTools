### The number next to P is in the original radar filename -- so below I provide an example of reading the file for P16

```python
#### Get file list
fns_radar = sorted(glob.glob('/mnt/data01/Data/PolarGap/radar/segy_4us/*16*.segy'))
```

```python
from segysak.segy import segy_loader, well_known_byte_locs, segy_header_scrape, segy_bin_scrape

########### This section reads the trace headers to determine the timing information
trace_headers = segy_header_scrape(fns_radar[0])
trace_headers.iloc[:,38:40]
```

```python
bin_header = segy_bin_scrape(fns_radar[0])
bin_header
```

```python
trace_headers.iloc[0,39]
```

```python
####################### This uses the trace header information to construct a time axis
cice = 168000000
time = np.arange(0,trace_headers.iloc[0,38])*trace_headers.iloc[0,39]/(10**9)
depth = (time).squeeze()/2*cice
```

```python
data = segy_loader(fns_radar[0], vert_domain="TWT")
data
```

```python
plt.figure(figsize=(20,10))
plt.imshow(
    np.transpose(10*np.log10(np.power(data['data'].values,2))),
    extent=[data['cdp'].min().values,data['cdp'].max().values,data['twt'].min().values,data['twt'].max().values],
    aspect='auto',
    cmap='gray',
    vmin=-25,
    vmax=50,
)
plt.colorbar()
```
